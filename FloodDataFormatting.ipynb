```python
import csv
with open('FloodData.csv','r') as flood_data:
    flood_data = csv.reader(flood_data, quoting=csv.QUOTE_NONE) #creates csv_reader object from which a list of data may be assembled
    flood_list = []
    for country in flood_data:
        flood_list.append(country) #stores all csv data in two by two list where first index is order and second index is order data
with open('GDPGrowth.csv','r') as gdp_data:
    gdp_data = csv.reader(gdp_data, quoting=csv.QUOTE_NONE) #creates csv_reader object from which a list of data may be assembled
    gdp_list = []
    for country in gdp_data:
        gdp_list.append(country)
```

```python
for country in flood_list:
    if country[0].startswith('"'):
        country[0] = country[0][1:]
for country in gdp_list:
    if country[0].startswith('"'):
        country[0] = country[0][1:]
```

```python
countryFound = False
for gdp_index in range(1, len(gdp_list) - 1):
    for flood_index in range(1, len(flood_list)):
        if gdp_list[gdp_index][0] == flood_list[flood_index][0]:
            countryFound = True
    if countryFound == False:
        blank_country[0] = gdp_list[gdp_index][0]
        flood_list.insert(gdp_index, blank_country.copy())
    countryFound = False
```

```python
dates = ['1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968', '1969']
filler = ['', '', '', '', '', '', '', '', '']
for date_index in range(0, len(dates)):
    flood_list[0].insert(1 + date_index, dates[date_index])
flood_list[0].insert(1, dates)
for country in flood_list:
    if country[0] != 'People affected by Flooding':
        for fill_index in range(0, len(filler)):
            country.insert(1 + fill_index, filler[fill_index])
```
